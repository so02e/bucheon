```python
import folium
import pandas as pd
import json

lat_lon = {'':[37.4851398,126.8116223],
           '':[37.484085,126.7806143],
           '':[37.487626,126.7508563],
           '':[37.482755,126.7934653],
           '':[37.485731,126.7638988],
           '':[37.5061657,126.8040656],
           '':[37.505452,126.7953093],
           '':[37.5041873,126.7847732],
           '':[37.503064,126.7737883],
           '':[37.504681,126.7614293],
           '':[37.505814,126.7509743]   
}

subway_bucheon = pd.DataFrame(lat_lon)
subway_bucheon = subway_bucheon.transpose()
subway_bucheon = subway_bucheon.reset_index()
subway_bucheon.columns = ['','lat','lon']

```

```python
#   
subway_map = folium.Map(location=[37.50554861215234,126.77550612183495], zoom_start=13,tiles='cartodbpositron')
geo_data = "data/HangJeongDong_ver20210101.geojson"

    
#  
for name, lat, lng in zip(subway_bucheon., subway_bucheon.lat, subway_bucheon.lon):
    folium.CircleMarker([lat, lng],
                        radius=5,         #  
                        fill=True,
                        fill_color='purple',    #   
                        fill_opacity=0.7, #     
                        popup=name
    ).add_to(subway_map)

display(subway_map)
```
