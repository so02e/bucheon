```python
# (, )
## (1)    
import pandas as pd
bus = pd.read_csv('./data/.csv')

is_bucheon = bus[''] == ''
bucheon_bus = bus[is_bucheon]


#   
import folium
bucheon_map = folium.Map(location=[37.50554861215234,126.77550612183495], tiles='Stamen Terrain', zoom_start=13)

# 
import json
geo_path = 'data/.json'
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))
    
#    
for name, lat, lng in zip(bucheon_bus., bucheon_bus.WGS84, bucheon_bus.WGS84):
    folium.CircleMarker([lat, lng],
                        radius=3,         #  
                        fill=True,
                        fill_color='purple',    #   
                        fill_opacity=0.7, #     
                        popup=name,
                        geo_data=geo_data,    #  
    ).add_to(bucheon_map)

display(bucheon_map)
```
